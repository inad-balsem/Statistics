### 3. Correlation between customer service call and profile total call

```python
profile_complain = ["high" if x>2 else "low" for x in df['number_customer_service_calls']]
df['profile_complain'] = profile_complain
df['profile_complain'].head(10)
```

```python
df_high_comp = df[df['profile_complain']=="high"]
df_low_comp = df[df['profile_complain']=="low"]
column_select = ['total_day_calls','total_eve_calls','total_night_calls']
```

```python
df_high_comp[column_select].mean()
```

```python
df_low_comp[column_select].mean()
```

Conclusion:
<br> there are almost no relation between customer service call and call activity

### 4. Profile country with the highest churn

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['state'].unique()
```

```python
df['class'].unique()
df_churn = df[df['class']==1]
```

```python
plt.figure(figsize=(10,10))
g = sns.countplot(x="state", data = df_churn, palette = 'Set1')
g.set_xticklabels(g.get_xticklabels(), rotation = 90, ha ="right")
g
plt.title('State VS Churn Profile', size = 20)
```

Conclusion 
<br> the highest churn is from state with code 31
